```python
import pandas as pd
from shapely.geometry import Point, LineString, shape, MultiLineString
import geopandas as gpd
import matplotlib.pyplot as plt
from geopandas import GeoDataFrame
from shapely import ops
```

```python
df = pd.read_excel("AeroviasS.xls")
df.head()
```

```python
points= df.apply(lambda row: Point(row.Longitud, row.Latitud), axis=1)
```

```python
radioayudas= gpd.GeoDataFrame(df, geometry=points)
radioayudas.crs= "WGS84"
radioayudas.head()
```

```python
radioayudas['Aerovia'].unique()
```

```python
condicionJ1= radioayudas[radioayudas["Aerovia"]=='J1']
condicionJ13= radioayudas[radioayudas["Aerovia"]=='J13']
condicionJ21= radioayudas[radioayudas["Aerovia"]=='J21']
condicionJ22= radioayudas[radioayudas["Aerovia"]=='J22']
condicionJ25= radioayudas[radioayudas["Aerovia"]=='J25']
condicionJ26= radioayudas[radioayudas["Aerovia"]=='J26']
condicionJ29= radioayudas[radioayudas["Aerovia"]=='J29']
condicionJ42= radioayudas[radioayudas["Aerovia"]=='J42']
condicionJ92= radioayudas[radioayudas["Aerovia"]=='J92']
condicionJ93= radioayudas[radioayudas["Aerovia"]=='J93']
condicionJ141= radioayudas[radioayudas["Aerovia"]=='J141']
condicionJ177= radioayudas[radioayudas["Aerovia"]=='J177']
```

```python
J1 = LineString(condicionJ1['geometry'])
J1.length
```

```python
J1
```

```python
geometry = [Point(xy) for xy in zip(df.Longitud, df.Latitud)]
df1 = GeoDataFrame(df, geometry=geometry)
# Aggregate these points with the GroupBy
df1 = df1.groupby(['Aerovia'])['geometry'].apply(lambda x: LineString(x.tolist())if x.size > 1 else x.tolist())
df1 = GeoDataFrame(df, geometry='geometry')
LineString(df1.geometry)
```

```python
mergedRoad = ops.linemerge(LineString(df1.geometry))
mergedRoad

mergedRoad.length
```

```python
# Crear ruta de salida
outfp = "Aerovias.shp"

# Escribir en archivo shapefile (Formato por defecto Shapefile)
df1.to_file(outfp)
```

```python
LineString(df1.geometry)
lxy = LineString(df1.geometry).xy
lxy
```

```python
plt.style.use('dark_background')
# Control del tamao de la figura del mapa
fig, ax = plt.subplots(figsize=(80, 50))
 
# Control del ttulo y los ejes
ax.set_title('Radio ayudas de la Republica Mxicana', 
             pad = 20, 
             fontdict={'fontsize':20, })
ax.set_xlabel('Longitud')
ax.set_ylabel('Latitud')
 
LineString(df1.geometry).plot(cmap='plasma', ax=ax, zorder=5)
```
